```python
xticks_pos = [1,10,100,1000,5000,10000]
xticks_label = [j/n for j in xticks_pos]
plt.figure(figsize=(14,10))
color_func = plt.get_cmap('tab10')
for i,block_size in enumerate(range(1,200,30)):
    Xs = list(range(n))
    Ys = [KbMOM.minimal_number_of_blocks(n=n,nb_outliers=k,block_size=block_size,alpha=alpha) for k in Xs]
    plt.plot(Xs,Ys,label=block_size,color=color_func(i))
    plt.vlines(x=Xs[argmax(Ys)-1],ymin=0,ymax=1e6,linestyles=':',color=color_func(i))

plt.title(
    f'''Plot of the evolution maximal possible number of blocks in KbMOM method in function of the outlier proportion \n
for several block sizes, so that the probability that the estimated mean is directly affected by the outliers is less than {alpha} \n
after a given proportion of outliers the minimal_number_of_blocks is not defined, hence these curves'''
          ,fontsize=15)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.xticks(xticks_pos,xticks_label)
plt.show()
```

## Remarks

Here, one can see how the two function to control the block sizes and the number of block work.
Note that the algorithmique complexity diverges with the number of blocks. So, watch that out and dont take the maximal allowed block size in order to get an output.
