# Show basics of TCLUST

### preamble

TClust is a robust clustering estimator that searches for the best mean vector and covariance matrices of a gaussian mixture model with maximum K components (you can set K=5 and only get 3 components if the model optimization procedure fits it that way).

/!\ it is currently impossible to set the random seed. Each new run of TClust will result in a different result.

The python version of TCLUST shown here is a version that has been carefully transcripted in python from the available code in language "R" (https://cran.r-project.org/web/packages/tclust/index.html in its version of 2019).

The code in R may have evolve since 2019 but the algorithme named "TClust" has not changed and is a well known algorithm in statistics.

## TCLUST

Tclust is a function (not a class).

It outputs a dictionary with all useful information. See the available keys.

```python
output = TCLUST(
    x=X, # data
    k=3,
    alpha = 0.05,
    restr_fact=12,
    m = 5,
    equal_weights = False,
    fuzzy = False, # necessary arguments for computation
    init_centers=None,
    init_sigma=None,
    init_cluster_size=None,
    init_cluster_weights=None, # control of the initialization
    nstart = 1,
    iter_max = 20,
    zero_tol = 1e-16,
    store_x = True,
    trace = None, # optional arguments to control some parameters, output and loops
#     f_restr = restr_diffax, # the default value is restr_diffax defined in the TCLUST.py file.
    restr_deter=None,
    f_hook_iter=None,
    f_hook_model=None # intern functions that could be used. Only f_restr was compulsory to implement to run tclust
)
```

```python
output.keys()
```

description of output objects:

- 'centers' : all final centers stored in an 2darray, each center can be obtained in the 2nd dimension (i.e. array[:,wanted_one] where "wanted_one" is an integer)

- 'cov' : all final covariance matrices stored in an 3darray, each matrix can be obtained in the 1st dimension (i.e. array[wanted_one] where "wanted_one" is an integer)

- 'cluster' : final assignment array indicating in position i which cluster the i-th data was assigned to.

-  'tclust_input' : a dictionary with all the inputs given to the function. This is here because it can be convienent to debug and also because the original code (in R) did that. In particular, it can be used to compare the maximal sought K components to the number of remaining components after optimization.

- 'k' : number of components at the end of the optimization process

- 'obj' : the final value of the objective function. refer yourself to the code or to the original mathematical article to see what is the objective function

- 'size' : array which indicates at position i the size of the i-th component

- 'weights' : array that indicates in position i the weights associated to the i-th component in the gaussian mixture model (often called pi_{i} or a_{i})

- 'ret_orig' : a dictionary of internal variables. TClust was coded in R according to a functional paradigm. For that reason, a dictionary named dict_iter was passed to all functions that are called in TClust and output with updates. 'ret_orig' is this dictionary. It does not contain any important informations that are not already mention above but we coded it that way in python to make it easier to compare the to implementation and to assert that that do exactly the same thing.

## visualize output

```python
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms

def covariance_ellipse(mean_vector, cov, ax, facecolor='none', **kwargs):
    
    mean_x, mean_y = mean_vector[0], mean_vector[1]
    u, s, vh = np.linalg.svd(a=cov)
    v1, v2 = u
    l1, l2 = s
    
    angle = np.arctan(v1[1]/v1[0]) * 360 / (2*np.pi)
    
    ellipse = Ellipse((mean_x, mean_y), width=2*l1**0.5 , height=2*l2**0.5 , angle=angle, facecolor=facecolor, **kwargs)
    return ax.add_patch(ellipse)
```

```python
color_func = plt.get_cmap('tab10')
fig, ax = plt.subplots(1,1,figsize=(15,15))
ax.scatter(X[:,0],X[:,1],c=output['cluster'],s=12,label='estimated clusters, plus estimated outliers in purple')
ax.scatter(output['centers'][0],output['centers'][1],c='red',label='estimated mean vector')
label_done = False
for center, cov in zip(output['centers'].T,output['cov']):
    if not label_done:
        covariance_ellipse(center, cov, ax, edgecolor='red',label='ellispe showing the covariance matrix of its cluter')
        label_done = True
    else:
        covariance_ellipse(center, cov, ax, edgecolor='red')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
