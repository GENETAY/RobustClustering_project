```python
# import different de ceux tout en haut pour ne rien toucher a ce qui marche
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

    
    
import numpy as np
import random
from collections import Counter

from RobustClustering import SMM, KbMOM

from sklearn.datasets import make_blobs
from sklearn.neighbors import LocalOutlierFactor

import matplotlib.pyplot as plt
%matplotlib inline
import time
```

```python
def def_centers(p,mu):
    # dimension de l'entree
    K,dim=mu.shape
    # les centres en dimension p
    centers=np.zeros((K,p))
    for k in range(K):
            for j in range(dim):
                centers[k,j]=mu[k,j]
    return(centers)

def data_clustered_outliers(n_samples,n_outliers,p,centers = np.array([[1,4],[2,1],[-2,3]]),centers_outliers=np.array([[7,7]]),it_=0):
    # initialisation des parametres
    centers=def_centers(mu=centers,p=p)
    centers_outliers=def_centers(mu=centers_outliers,p=p)
    K           = len(centers)              # nombre de classes
    sig         = 0.6                       # within variance
    # Generation des donnees
    Xinlier, y_inlier = make_blobs(n_samples=n_samples, centers=centers, cluster_std=0.60, random_state=it_)
    # Outliers creation
    Xoutlier, y_outlier = make_blobs(n_samples=n_outliers, centers=centers_outliers, cluster_std=0.60)
    # all concatenante
    n_data=n_samples+n_outliers
    X = np.concatenate([Xoutlier,Xinlier], axis=0)
    y_true=np.concatenate([y_outlier+K,y_inlier])
    return({'points':X,'partition':y_true,"centers":centers})
```

```python
n_samples=200
k=3
n_outliers=10
alpha=n_outliers/n_samples;print("alpha=",alpha)
```

```python
temp=data_clustered_outliers(n_samples=n_samples,n_outliers=n_outliers,p=2)
print(temp['partition'])
X=temp['points']
plt.plot(X[:,0],X[:,1],'b.')
```

one can see that the cluster "3" corresponds to the first data in the array.

# SMM in action

### /!\ SMM in our package must be initialized, otherwise there is an error

we did the manual because the SMM from sklearn that we copied and modified for the purpose of our numerical experiment could not be initialized. Moreover, it was always automatically initialized with Kmeans which is NOT a robust initialization.

```python
output_init = KbMOM(X=X,K=k,block_size=20, number_of_blocks=100,max_iter=0).fit(X=X)
```

```python
initial_centers = output_init['centroids']
initial_centers
```

```python
smm = SMM(n_components=k,random_state=1,init_center_given=initial_centers)
```

```python
output = smm.fit(X)
```

```python
output.predict(X)
```

```python
output.means
```

```python
color_func = plt.get_cmap('tab10')
fig, ax = plt.subplots(1,1,figsize=(15,15))
ax.scatter(X[:,0],X[:,1],c=output.predict(X),s=50,label='estimated clusters, plus estimated outliers in purple')
ax.scatter(output.means[:,0],output.means[:,1],c='red',label='estimated mean vector')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
