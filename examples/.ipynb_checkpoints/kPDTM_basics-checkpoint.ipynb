```python
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms

def covariance_ellipse(mean_vector, cov, ax, size_scale=1, facecolor='none', **kwargs):
    
    mean_x, mean_y = mean_vector[0], mean_vector[1]
    u, s, vh = np.linalg.svd(a=cov)
    v1, v2 = u
    l1, l2 = s
    
    angle = np.arctan(v1[1]/v1[0]) * 360 / (2*np.pi)
    
    ellipse = Ellipse((mean_x, mean_y), width=size_scale*2*l1**0.5 , height=size_scale*2*l2**0.5 , angle=angle, facecolor=facecolor, **kwargs)
    return ax.add_patch(ellipse)
```

## Make data

```python
from sklearn.datasets import make_blobs
```

```python
n_intlier = 1000
dim = 2
nb_outliers_around_inlier = 79
nb_outliers_far_from_inlier = 13
random_state = 665548366
K_true = 9
```

```python
X_in,y_in,centers = make_blobs(n_samples=n_intlier,n_features=dim,centers=K_true,return_centers=True,random_state=random_state)
X_out_1,y_out_1 = make_blobs(n_samples=nb_outliers_around_inlier,n_features=dim,centers=1,random_state=int(random_state/2))
X_out_2,y_out_2 = make_blobs(n_samples=nb_outliers_far_from_inlier,n_features=dim,centers=1,random_state=int(random_state/4))
X = np.concatenate([X_in,5*(X_out_1-X_out_1.mean(axis=0))+X_in.mean(axis=0),50*(X_out_2-X_out_2.mean(axis=0))+X_in.mean(axis=0)],axis=0)
y_in  = np.tile(A=0,reps=len(y_in))
y_out_1 = np.tile(A=max(y_in)+1,reps=len(y_out_1))
y_out_2 = np.tile(A=max(y_in)+1,reps=len(y_out_2))
y = np.concatenate([y_in,y_out_1,y_out_2])
```

how many datapoints?

```python
n,_ = X.shape
n
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(20,6))
color_func = plt.get_cmap('tab10')
colors = [color_func(cluster) for cluster in y]
# fig 1: all data
ax[0].scatter(X[:,0],X[:,1],c=colors,s=4,label='data (blue), outliers (orange)')
ax[0].scatter(centers[:,0],centers[:,1],c='red',label='true centroids')
ax[0].legend()
# fig 2: zoom
ax[1].scatter(X[:,0],X[:,1],c=colors,s=4,label='data (blue), outliers (orange)')
ax[1].scatter(centers[:,0],centers[:,1],c='red',label='true centroids')
ax[1].set_xlim(-15,10)
ax[1].set_ylim(-15,10)
fig.show()
```

# kPDTM and kPLM in action

## kPDTM

```python
q = 50
k = 10
sig = 1000 # sig = signal. it is the number of data to take into account. in other terms, the inliers.
iter_max = 100
nstart = 1
kPDTM_values, centroids, means, variances, colors, cost = kPDTM(X,X,q,k,sig,iter_max,nstart)  
```

```python
centers
```

```python
fig, ax = plt.subplots(1,1,figsize=(9,9))
color_func = plt.get_cmap('tab10')
ax.scatter(X[:,0],X[:,1],s=8,label='data',c=colors)
ax.scatter(centroids[:,0],centroids[:,1],s=32,c='black',label='centroids at the end of the kPDTM iterations')
label_done = False
for mu in centroids:
    cov = np.eye(2)
    if not label_done:
        covariance_ellipse(mu, cov, ax, edgecolor='red',label='ellispe showing the covariance matrix of its cluter')
        label_done = True
    else:
        covariance_ellipse(mu, cov, ax, edgecolor='red')
ax.legend()
ax.set_xlim(-15,10)
ax.set_ylim(-15,10)
plt.show()
```

## kPLM

```python
%%time

q = 50
k = 10
sig = 1000
iter_max = 30
nstart = 1
kPLM_values, centroids, Sigma, means, weights, colors, cost = kPLM(X,X,q,k,sig,iter_max,nstart) 
```
